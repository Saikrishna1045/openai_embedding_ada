```python
import openai
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from langchain.embeddings import OpenAIEmbeddings
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os 
openai_api_key = os.getenv("OPENAI_API_KEY")
```

```python
# Define the text samples
text1 = "Artificial intelligence is transforming industries."
text2 = "Machine learning and AI are revolutionizing the tech world."
```

```python
# Function to generate text embeddings using 'text-embedding-ada-002'
def get_embedding(text, model="text-embedding-ada-002"):
    response = openai.Embedding.create(
        input=text,
        model=model
    )
    return response['data'][0]['embedding']
```

```python
# Generate embeddings for both texts
embedding1 = get_embedding(text1)
embedding2 = get_embedding(text2)
```

```python
print({"1": embedding1})
print({"2": embedding2})
```

```python
# Convert the embeddings to numpy arrays for easier manipulation
embedding1_np = np.array(embedding1).reshape(1, -1)
embedding2_np = np.array(embedding2).reshape(1, -1)
```

```python
print({"1": embedding1_np})
print({"2": embedding2_np})
```

```python
# Compute cosine similarity between the two embeddings
similarity = cosine_similarity(embedding1_np, embedding2_np)
```

```python
print(f"Cosine Similarity between the two texts: {similarity[0][0]:.4f}")
```

Cosine Similarity between the two texts: 0.9234
